```python
import pandas as pd
import seaborn as sns

vectores_df = pd.read_csv('vectores.csv')
labels_df = pd.read_csv('labels.csv')

vectores_df['label'] = labels_df[list(labels_df)[0]]
```

```python
[x for x in dir(sns) if 'plot' in x]
```

```python
%matplotlib inline
sns.scatterplot(data=vectores_df, x='x', y='y', hue='label', alpha=0.03)
```

```python
under_0 = vectores_df[vectores_df.label==0]
sns.scatterplot(data=under_0, x='x', y='y', hue='label')
```

```python
under_1 = vectores_df[vectores_df.label==1][vectores_df.x<0.3]
sns.scatterplot(data=under_1, x='x', y='y', hue='label')
```
