```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('../results/knn_cuantitative_reduced.csv')
df = df.groupby(['k'],as_index=False)['time'].mean()
df = df[1:]
df['time'] /= 10**9
plotname = 'knn_cuantitative_reduced:'
df.plot(x='k',y='time', title=plotname)
```

```python
df = pd.read_csv('../results/knn_cuantitative_big.csv')
df = df.groupby(['k'],as_index=False)['time'].mean()
df['time'] /= 10**9
plotname = 'knn_cuantitative_big:'
df.plot(x='k',y='time', title=plotname)
```

```python
import matplotlib.pyplot as plt
def plot_time_vs_alpha(df, rows, save=False):
    df = df[df['rows'] == rows]
    df['time'] /= 10**9
    plotname = 'pca_cuantitative_redImg_{}_rows'.format(rows)
    df.plot.scatter(x='alpha',y='time', title=plotname)
    if save:
        plt.savefig('./plots/{}.png'.format(plotname))
        
```

```python
df = pd.read_csv('../results/pca_cuantitative_reduced.csv')
for row_val in sorted(list(set(df['rows'])), key=int):
    plot_time_vs_alpha(df, row_val,True)
```

```python
df = pd.read_csv('../results/pca_cuantitative_big.csv')

for row_val in sorted(list(set(df['rows'])), key=int):
    plot_time_vs_alpha(df, row_val)
```

```python
df = pd.read_csv('../results/pca_knn_cuantitative_reduced.csv')
df = df.groupby(['k'],as_index=False)['time'].mean()
df['time'] /= 10**9
plotname = 'pca_knn_cuantitative_reduced'
df.plot(x='k',y='time', title=plotname)
```

```python
df = pd.read_csv('../results/pca_knn_cuantitative_reduced_200_reps.csv')

for alpha_val in sorted(list(set(df['alpha'])), key=int):
    transient_df = df[df['alpha'] == alpha_val]
    transient_df = transient_df.groupby(['k'],as_index=False)['time'].mean()
    transient_df['time'] /= 10**9
    plotname = 'pca_knn_cuantitative_reduced, alpha = {}'.format(alpha_val)
    transient_df.plot.scatter(x='k',y='time', title=plotname)
```
